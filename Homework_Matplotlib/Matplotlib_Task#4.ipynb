Данный датасет является примером несбалансированных данных, так как мошеннические операции с картами встречаются реже обычных.

Импортируйте библиотеку Pandas, а также используйте для графиков стиль “fivethirtyeight”.

Посчитайте с помощью метода value_counts количество наблюдений для каждого значения целевой переменной Class и примените к полученным данным метод plot, чтобы построить столбчатую диаграмму. Затем постройте такую же диаграмму, используя логарифмический масштаб.

```python
import pandas as pd
from matplotlib import pyplot as plt
```

```python
plt.style.use("fivethirtyeight")
```

```python
fraud = pd.read_csv("creditcard.csv")
```

```python
fraud["Class"].value_counts()
```

```python
fraud["Class"].value_counts().plot(kind="bar")
```

```python
fraud["Class"].value_counts().plot(kind="bar", logy=True)
```

```python
plt.hist(fraud[fraud["Class"] == 0]["V1"], density=True, bins=20, alpha=0.5, color="grey", label="Class 0")
plt.hist(fraud[fraud["Class"] == 1]["V1"], density=True, bins=20, alpha=0.5, color="red", label="Class 1")
plt.xlabel("Class")
plt.legend(loc="upper left")

plt.show()
```

Альтернативный вариант, исключающий наложение гистограмм друг на друга.

```python
plt.hist([fraud[fraud["Class"] == 0]["V1"],fraud[fraud["Class"] == 1]["V1"]], density=True, bins=20, alpha=0.5, 
         label=["Class 0", "Class 1"])
plt.xlabel("Class")
plt.legend(loc="upper left")

plt.show()
```
